# Change figure color
## 1. Setting color first
Using the parameter `c`

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 1, 0.1)
ys = [np.random.rand(len(x)) for i in range(5)]

# the colors length should the same with the kind of data you input 
colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:cyan']
fig, ax = plt.subplots(figsize=(10,8))

for i in range(len(ys)):
    ax.scatter(x, ys[i], label=f'$y_{i}$', marker='.', c=colors[i], s=100)
    
ax.legend(loc='upper left')


plt.show()
```

## 2. Change color later
Using `.collections[i].set_color()`

```python
x = np.arange(0, 1, 0.1)
ys = [np.random.rand(len(x)) for i in range(5)]

# the colors length should the same with the kind of data you input 

fig, ax = plt.subplots(figsize=(10,8))

for i in range(len(ys)):
    ax.scatter(x, ys[i], label=f'$y_{i}$', marker='.', s=100)

# setting color after plotting function
ax.collections[0].set_color('tab:red')
ax.collections[1].set_color('tab:blue')
ax.legend(loc='upper left')

plt.show()
```
